**Cel zajęć:**  
Celem zajęć jest zapoznanie się z iteracyjnymi metodami rozwiązywania układów równań liniowych w postaci macierzowej. Metody iteracyjne, w przeciwieństwie do analitycznych, nie zwracają dokładnego rozwiązania, a jedynie jego przybliżenie. Przy spełnionych kryteriach, przybliżenia obliczane w kolejnych iteracjach są zbieżne do dokładnego rozwiązania. Metody iteracyjne znajdują swoje zastosowanie szczególnie przy dużych układach równań postaci: 
   
$
\mathbf{Ax}=\mathbf{b}
$ 
  
gdzie $\mathbf{A}$ to macierz współczynników z lewej strony równania, wektor $\mathbf{x}$ jest wektorem zmiennych a wektor $\mathbf{b}$ wyników prawej strony równania.  

Do oceny jakości rozwiązania będziemy wykorzystywać residuum (ang. *residual*)
$
\mathbf{r}=\mathbf{b}-\mathbf{Ax}
$

***Zadanie 5***   
Rozwiąż układ równań z Zadania 3 podpunkt a) przy pomocy:

- funkcji *[solve](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.linalg.solve.html?highlight=solve#numpy.linalg.solve)* z biblioteki numpy

- zaimplementowanej iteracyjnej metody *solve_jacobi*

- iteracyjnej metody *[gradientów sprzężonych](https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.linalg.cg.html)*
 z biblioteki scipy

- iteracyjnej metody *[GMRES](https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.linalg.gmres.html)*
 z biblioteki scipy

Przed każdą metodą sprawdź czy warunki dotyczące macierzy dla danej metody są spełnione. Zbadaj dokładność i czas obliczeń metod w zależności od liczby iteracji. Wyniki przedstaw na wykresach.

```python
maxiters = [10, 25,  50, 75,  100]
labels = [str(el) for el in maxiters]

x_solve = []
res_solve = []
times_solve = []

x_jac = []
iter_jac = []
res_jac = []
times_jac = []

x_grad = []
iter_grad = []
res_grad = []
times_grad = []

x_gmres = []
iter_gmres = []
res_gmres = []
times_gmres = []

for maxiter in maxiters:   
    x_solve_curr = np_linalg.solve(A1, b1)
    x_solve.append(x_solve_curr)
    res_solve.append(main.residual_norm(A1, x_solve_curr, b1))
    time_solve = %timeit -r 5 -n 10 -o np_linalg.solve(A1, b1)
    times_solve.append(time_solve.average)  

    x_jac_curr, iter_jac_curr = main.solve_jacobi(A1, b1, np.zeros((A1.shape[0],)), maxiter= maxiter)
    x_jac.append(x_jac_curr)
    iter_jac.append(iter_jac_curr)
    res_jac.append(main.residual_norm(A1, x_jac_curr, b1))
    time_jac = %timeit -r 5 -n 10 -o main.solve_jacobi(A1, b1, np.zeros(A1.shape[0],), maxiter= maxiter)
    times_jac.append(time_jac.average)

    x_grad_curr, iter_grad_curr = sp_linalg.cg(A1, b1, maxiter= maxiter)
    x_grad.append(x_grad_curr)
    iter_grad.append(iter_grad_curr)
    res_grad.append(main.residual_norm(A1, x_grad_curr, b1))
    time_grad = %timeit -r 5 -n 10 -o sp_linalg.cg(A1, b1, maxiter= maxiter)
    times_grad.append(time_grad.average)

    x_gmres_curr, iter_gmres_curr = sp_linalg.gmres(A1, b1, maxiter= maxiter)
    x_gmres.append(x_gmres_curr)
    iter_gmres.append(iter_gmres_curr)
    res_gmres.append(main.residual_norm(A1, x_gmres_curr, b1))
    time_gmres = %timeit -r 5 -n 10 -o sp_linalg.gmres(A1, b1, maxiter= maxiter)
    times_gmres.append(time_gmres.average)

_, ((ax7, ax8), (ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(4, 2, figsize= (14, 20))
ax7.plot(labels, res_solve, 'o', color= 'teal')
ax7.set(title= 'Dokładność funkcji $solve$', xlabel= 'Liczba iteracji', ylabel= 'Norma residuum')
ax7.grid()
ax8.plot(labels, times_solve, 'o', color= 'rebeccapurple')
ax8.set(title= 'Czas obliczeń funkcji $solve$', xlabel= 'Liczba iteracji', ylabel= 'Czas [s]')
ax8.grid()

ax1.plot(labels, res_jac, 'o', color= 'teal')
ax1.set(title= 'Dokładność metody $solve_ jacobi$', xlabel= 'Liczba iteracji', ylabel= 'Norma residuum')
ax1.grid()
ax1.semilogy()
ax2.plot(labels, times_jac, 'o', color= 'rebeccapurple')
ax2.set(title= 'Czas obliczeń metody $solve_ jacobi$', xlabel= 'Liczba iteracji', ylabel= 'Czas [s]')
ax2.grid()

ax3.plot(labels, res_grad, 'o', color= 'teal')
ax3.set(title= 'Dokładność metody gradientów sprzężonych', xlabel= 'Liczba iteracji', ylabel= 'Norma residuum')
ax3.grid()
ax3.semilogy()
ax4.plot(labels, times_grad, 'o', color= 'rebeccapurple')
ax4.set(title= 'Czas obliczeń metody gradientów sprzężonych', xlabel= 'Liczba iteracji', ylabel= 'Czas [s]')
ax4.grid()

ax5.plot(labels, res_gmres, 'o', color= 'teal')
ax5.set(title= 'Dokładność metody GMRES', xlabel= 'Liczba iteracji', ylabel= 'Norma residuum')
ax5.grid()
ax6.plot(labels, times_gmres, 'o', color= 'rebeccapurple')
ax6.set(title= 'Czas obliczeń metody GMRES', xlabel= 'Liczba iteracji', ylabel= 'Czas [s]')
ax6.grid()

_, ax = plt.subplots(figsize= (8, 6))
ax.plot(labels, res_solve, 'o', color= 'silver', label= '$solve$', markersize= 16 )
ax.plot(labels, res_jac, 'o', color= 'salmon', label= '$solve_ jacobi$', markersize= 12)
ax.plot(labels, res_grad, 'o', color= 'cornflowerblue', label= 'gradienty sprzężone', markersize= 8)
ax.plot(labels, res_gmres, 'o', color= 'purple', label= 'GMRES', markersize= 4)
ax.legend()
ax.grid()
ax.semilogy()
ax.set(title= 'Porównanie dokładności metod', xlabel= 'Liczba iteracji', ylabel= 'Norma residuum')


```

Najmniej dokładne okazały się funkcje wykorzystujące metody iteracyjne gradientów sprzężonych oraz *GMRES*.
